```python
#Importing necessary libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, Dense, Flatten,
                                     Dropout, Add, Activation, BatchNormalization,
                                     GlobalAveragePooling2D, Concatenate, AveragePooling2D)
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist, cifar10, cifar100
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import datetime
import os
import cv2
import gc
import tensorflow.keras as keras
from PIL import Image
import math
from sklearn.metrics import confusion_matrix


#Define a function to build a VGG model
def Building_VGG_Model(input_shape, num_classes):
    """Builds a Sequential VGG-like model for image classification.

    Args:
        input_shape (tuple): Shape of the input image data.
        num_classes (int): Number of classes in the dataset.

    Returns:
        model (Sequential): A Keras model instance compiled with Adam optimizer.
    """
    model = Sequential()
    #Layer Block 1: Convolutional + MaxPooling
    model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape))
    model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # Add more layers if required for deeper networks
    #Flatten and Dense layers to classify images
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    #Compiling th model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return model



def Building_Resnet_Model(input_shape, num_classes):
    """Builds a ResNet model using the functional API with residual blocks.

    Args:
        input_shape(tuple): Shape of the input image data.
        num_classes(int): Number of classes in the dataset.

    Returns:
        model (Model): A Keras Model instance compiled with Adam optimizer.
    """
    def residual_block(x, filters, kernel_size=(3, 3), strides=(1, 1), use_conv_shortcut=False):
        """Adds a residual block with optional shortcut convolution.

        Args:
            x (tensor): Input tensor to the residual block
            filters (int): Number of filters for the convolutional layers
            kernel_size (tuple): Size of the kernel for the convolutional layers
            strides (tuple): Strides for the convolution
            use_conv_shortcut (bool): Whether to include a convolution in the shortcut

        Returns:
            x (tensor): Output tensor from the block.
        """
        shortcut = x
        if use_conv_shortcut:
            shortcut = Conv2D(filters, kernel_size=(1, 1), strides=strides,
                              kernel_regularizer=l2(1e-4))(x)
            shortcut = BatchNormalization()(shortcut)

        x = Conv2D(filters, kernel_size, padding='same', strides=strides, kernel_regularizer=l2(1e-4))(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)

        x = Conv2D(filters, kernel_size, padding='same', kernel_regularizer=l2(1e-4))(x)
        x = BatchNormalization()(x)

        x = Add()([shortcut, x])
        x = Activation('relu')(x)
        return x

    inputs = Input(shape=input_shape)
    x = Conv2D(64, (7, 7), strides=(2, 2), padding='same', kernel_regularizer=l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)
   
    #Building residual blocks
    x = residual_block(x, 64)
    x = residual_block(x, 64)
    
    #Increase number of filters with each block
    x = residual_block(x, 128, strides=(2, 2), use_conv_shortcut=True)
    x = residual_block(x, 128)

    x = residual_block(x, 256, strides=(2, 2), use_conv_shortcut=True)
    x = residual_block(x, 256)

    x = residual_block(x, 512, strides=(2, 2), use_conv_shortcut=True)
    x = residual_block(x, 512)

    x = GlobalAveragePooling2D()(x)
    x = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=x)
    model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

    return model


#Define a function to build a GoogleNet model
def Building_Googlenet_Model(input_shape, num_classes):
    """Builds a GoogleNet model with Inception modules using the functional API.

    Args:
        input_shape (tuple): Shape of the input image data.
        num_classes (int): Number of classes in the dataset.

    Returns:
        model (Model): A Keras Model instance compiled with Adam optimizer.
    """
    def Inception_module(x, filters_1x1, filters_3x3_reduce, filters_3x3, filters_5x5_reduce, filters_5x5, filters_pool_proj):
        """Constructs an Inception module.

        Args:
            x (tensor): Input tensor to the Inception module.
            filters_1x1 (int): Number of filters for the 1x1 convolution.
            filters_3x3_reduce (int): Number of filters to reduce dimension before the 3x3 convolution.
            filters_3x3 (int): Number of filters for the 3x3 convolution.
            filters_5x5_reduce (int): Number of filters to reduce dimension before the 5x5 convolution.
            filters_5x5 (int): Number of filters for the 5x5 convolution.
            filters_pool_proj (int): Number of filters for the projection after pooling.

        Returns:
            output (tensor): Concatenated output from all convolutions and pooling in the module.
        """
        conv_1x1 = Conv2D(filters_1x1, (1, 1), padding='same', activation='relu')(x)

        conv_3x3_reduce = Conv2D(filters_3x3_reduce, (1, 1), padding='same', activation='relu')(x)
        conv_3x3 = Conv2D(filters_3x3, (3, 3), padding='same', activation='relu')(conv_3x3_reduce)

        conv_5x5_reduce = Conv2D(filters_5x5_reduce, (1, 1), padding='same', activation='relu')(x)
        conv_5x5 = Conv2D(filters_5x5, (5, 5), padding='same', activation='relu')(conv_5x5_reduce)

        pool = MaxPooling2D((3, 3), strides=(1, 1), padding='same')(x)
        pool_proj = Conv2D(filters_pool_proj, (1, 1), padding='same', activation='relu')(pool)

        output = Concatenate(axis=-1)([conv_1x1, conv_3x3, conv_5x5, pool_proj])
        return output
    
    
    input_layer = Input(shape=input_shape)

    #Initial convolution and pooling layers
    x = Conv2D(64, (7, 7), strides=(2, 2), padding='same', activation='relu')(input_layer)
    x = MaxPooling2D((3, 3), padding='same', strides=(2, 2))(x)
    x = Conv2D(64, (1, 1), padding='same', activation='relu')(x)
    x = Conv2D(192, (3, 3), padding='same', activation='relu')(x)
    x = MaxPooling2D((3, 3), padding='same', strides=(2, 2))(x)
    
    #Adding 2 Inception modules
    x = Inception_module(x, filters_1x1=64, filters_3x3_reduce=96, filters_3x3=128,
                         filters_5x5_reduce=16, filters_5x5=32, filters_pool_proj=32)

    x = Inception_module(x, filters_1x1=128, filters_3x3_reduce=128, filters_3x3=192,
                         filters_5x5_reduce=32, filters_5x5=96, filters_pool_proj=64)

    x = MaxPooling2D((3, 3), padding='same', strides=(2, 2))(x)

    #Final Pooling and dense layer
    x = AveragePooling2D((2, 2), strides=(1, 1))(x)
    x = Flatten()(x)
    x = Dropout(0.4)(x)
    x = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=input_layer, outputs=x)
    model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

    return model


#Define a function to preprocess the data
def Data_preprocessing(x_train, x_test, target_shape=(32, 32, 3)):
    """Preprocesses the training and test images by resizing and normalizing them.

    Args:
        x_train (array): Training images.
        x_test (array): Test images.
        target_shape (tuple): Target size for the images.

    Returns:
        x_train, x_test (tuple of arrays): Preprocessed training and test image data.
    """
    x_train = x_train.astype('float32')
    x_test = x_test.astype('float32')
    x_train /= 255
    x_test /= 255

    x_train_resized = []
    x_test_resized = []

    for img in x_train:
        img_resized = cv2.resize(img, target_shape[:2])
        img_resized = np.stack([img_resized] * 3, axis=-1)  # Convert to 3 channels
        x_train_resized.append(img_resized)

    for img in x_test:
        img_resized = cv2.resize(img, target_shape[:2])
        img_resized = np.stack([img_resized] * 3, axis=-1)  # Convert to 3 channels
        x_test_resized.append(img_resized)

    x_train = np.array(x_train_resized)
    x_test = np.array(x_test_resized)

    return x_train, x_test

#Define a function to load and preprocess the CIFAR-10 dataset
def Load_and_preprocess_extra_dataset():
    """Loads and preprocesses the CIFAR-10 dataset by normalizing and one-hot encoding labels.

    Returns:
        x_train_extra, y_train_extra, x_test_extra, y_test_extra (tuple of arrays): Preprocessed training and test data.
    """
    (x_train_extra, y_train_extra), (x_test_extra, y_test_extra) = cifar10.load_data()

    x_train_extra = x_train_extra.astype('float32') / 255
    x_test_extra = x_test_extra.astype('float32') / 255

    y_train_extra = to_categorical(y_train_extra, 10)
    y_test_extra = to_categorical(y_test_extra, 10)

    return x_train_extra, y_train_extra, x_test_extra, y_test_extra

#Define a function to train the model
def Train_model(model, data_gen, x_train, y_train, x_test, y_test, dataset_name, model_name, epochs=100, batch_size=64):
    """Trains the model using the provided data generator and datasets.

    Args:
        model (Model): The Keras model to be trained.
        data_gen (ImageDataGenerator): Generator for data augmentation.
        x_train (array): Training data.
        y_train (array): Training labels.
        x_test (array): Test data.
        y_test (array): Test labels.
        dataset_name (str): Name of the dataset being used.
        model_name (str): Name of the model architecture.
        epochs (int): Number of training epochs.
        batch_size (int): Batch size for training.

    Returns:
        model (Model): Trained model.
        history (History): Training history object.
        start_time, end_time (datetime): Start and end times of the training.
    """
    start_time = datetime.datetime.now()
    print(f"Model: {model_name}, Dataset: {dataset_name}, Training started at: {start_time.strftime('%Y-%m-%d %H:%M:%S')}")

    history = model.fit(data_gen.flow(x_train, y_train, batch_size=batch_size),
                        steps_per_epoch=len(x_train) // batch_size,
                        epochs=epochs,
                        validation_data=(x_test, y_test),
                        verbose=1)

    end_time = datetime.datetime.now()
    print(f"Model: {model_name}, Dataset: {dataset_name}, Training finished at: {end_time.strftime('%Y-%m-%d %H:%M:%S')}")

    # Evaluate the model on the test dataset
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=1)

    # Print the test accuracy along with model name and dataset name
    print(f"Model: {model_name}, Dataset: {dataset_name}, Test Accuracy: {test_accuracy:.4f}")

    return model, history, start_time, end_time

#Define a function to save the training logs
def Save_logs(model_name, history, start_time, end_time):
    """Saves training logs to a file.

    Args:
        model_name (str): Name of the model.
        history (History): Training history object.
        start_time (datetime): Training start time.
        end_time (datetime): Training end time.
    """
    log_dir = os.path.join("logs", model_name)
    os.makedirs(log_dir, exist_ok=True)
    with open(os.path.join(log_dir, "history.txt"), "w") as f:
        f.write(f"Training started at: {start_time.strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Training finished at: {end_time.strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write("Training History:\n")
        f.write(str(history))

#Define a function to save the classified images
def Save_classified_images(model, x_test, y_test, dataset_name, network_name, num_images=10):
    """Saves a grid of classified images, showing true vs. predicted labels.

    Args:
        model (Model): The trained Keras model.
        x_test (array): Test data.
        y_test (array): Test labels.
        dataset_name (str): Name of the dataset.
        network_name (str): Name of the network architecture.
        num_images (int): Number of images to display in the grid.
    """
    # Get predictions
    y_prediction = model.predict(x_test)

    # Get the indices of the highest probability predictions
    y_prediction_labels = np.argmax(y_prediction, axis=1)
    y_true_labels = np.argmax(y_test, axis=1)

    # Create a directory to save the images if it does not exist
    base_dir = 'classified_images'
    dataset_dir = os.path.join(base_dir, dataset_name)
    network_dir = os.path.join(dataset_dir, network_name)

    if not os.path.exists(network_dir):
        os.makedirs(network_dir)

    # Set up grid dimensions
    grid_dim = int(math.sqrt(num_images))
    if grid_dim * grid_dim < num_images:
        grid_dim += 1

    # Create a grid of images
    fig, axes = plt.subplots(grid_dim, grid_dim, figsize=(15, 15))

    for i in range(grid_dim * grid_dim):
        ax = axes[i // grid_dim, i % grid_dim]
        if i < num_images:
            ax.imshow(x_test[i], cmap='gray')
            ax.set_title(f"True: {y_true_labels[i]} - Pred: {y_prediction_labels[i]}")
        ax.axis('off')

    # Save the grid as a single image
    plt.savefig(os.path.join(network_dir, f'classified_images_grid.png'), dpi=300)

    # Close the plot
    plt.close()




def Visualizing_Results(vgg_history, resnet_history, googlenet_history):
    """Plots the validation accuracy of VGG, ResNet, and GoogleNet models over training epochs.

    Args:
        vgg_history (History): Training history for VGG model.
        resnet_history (History): Training history for ResNet model.
        googlenet_history (History): Training history for GoogleNet model.
    """
    plt.figure(figsize=(12, 8))
    plt.plot(vgg_history.history['val_accuracy'], 'r', label='VGG', linewidth=2, marker='o', markersize=5)
    plt.plot(resnet_history.history['val_accuracy'], 'b', label='ResNet', linewidth=2, marker='o', markersize=5)
    plt.plot(googlenet_history.history['val_accuracy'], 'g', label='GoogleNet', linewidth=2, marker='o', markersize=5)
    plt.title('Validation Accuracy Comparison', fontsize=16, fontweight='bold')
    plt.xlabel('Epochs', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.legend(loc='upper left', fontsize=12)
    plt.grid(True, linestyle='-', linewidth=0.5)  # Solid grid lines for better visibility
    plt.tight_layout()
    plt.show()





#Define a function to evaluate the MNIST dataset
def Evaluate_mnist():
    """Loads the MNIST dataset, preprocesses it, trains and evaluates the models, and saves results.

    Returns:
        vgg_model, resnet_model, googlenet_model (tuple of Models): Trained Keras models.
    """
    #Load and preprocess MNIST data
    (x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = mnist.load_data()
    x_train_mnist, x_test_mnist = Data_preprocessing(x_train_mnist, x_test_mnist)
    data_gen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)
    y_train_mnist = to_categorical(y_train_mnist, 10)
    y_test_mnist = to_categorical(y_test_mnist, 10)

    # Train and evaluate models on MNIST
    vgg_model = Building_VGG_Model(input_shape=(32, 32, 3), num_classes=10)
    resnet_model = Building_Resnet_Model(input_shape=(32, 32, 3), num_classes=10)
    googlenet_model = Building_Googlenet_Model(input_shape=(32, 32, 3), num_classes=10)

    vgg_model,vgg_history,vgg_start_time, vgg_end_time = Train_model(vgg_model, data_gen, x_train_mnist, y_train_mnist, x_test_mnist, y_test_mnist,"mnist","vgg")
    resnet_model, resnet_history,resnet_start_time, resnet_end_time = Train_model(resnet_model, data_gen, x_train_mnist, y_train_mnist, x_test_mnist, y_test_mnist,"mnist","resnet")
    googlenet_model, googlenet_history,googlenet_start_time, googlenet_end_time = Train_model(googlenet_model, data_gen, x_train_mnist, y_train_mnist, x_test_mnist, y_test_mnist,"mnist","googlenet")

    Save_classified_images(vgg_model, x_test_mnist, y_test_mnist, "mnist", "vgg")
    Save_classified_images(resnet_model, x_test_mnist, y_test_mnist, "mnist", "resnet")
    Save_classified_images(googlenet_model, x_test_mnist, y_test_mnist, "mnist", "googlenet")

    # Save logs and visualize results
    Save_logs("mnist_vgg_logs.txt", vgg_history.history,vgg_start_time, vgg_end_time)
    Save_logs("mnist_resnet_logs.txt", resnet_history.history,resnet_start_time, resnet_end_time)
    Save_logs("mnist_googlenet_logs.txt", googlenet_history.history, googlenet_start_time, googlenet_end_time)

    Visualizing_Results(vgg_history, resnet_history, googlenet_history)
    return vgg_model, resnet_model, googlenet_model


#Define a function to evaluate to CIFAR-10 dataset
def Evaluate_cifar():
    """Loads the CIFAR-10 dataset, preprocesses it, trains and evaluates the models, and saves results.

    Returns:
        vgg_model, resnet_model, googlenet_model (tuple of Models): Trained Keras models.
    """
    # Load and preprocess CIFAR-10 data
    x_train_extra, y_train_extra, x_test_extra, y_test_extra = Load_and_preprocess_extra_dataset()
    data_gen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)

    # Train and evaluate models on CIFAR-10
    vgg_model = Building_VGG_Model(input_shape=(32, 32, 3), num_classes=10)
    resnet_model = Building_Resnet_Model(input_shape=(32, 32, 3), num_classes=10)
    googlenet_model = Building_Googlenet_Model(input_shape=(32, 32, 3), num_classes=10)

    vgg_model,vgg_history,vgg_start_time, vgg_end_time = Train_model(vgg_model, data_gen, x_train_extra, y_train_extra, x_test_extra, y_test_extra,"cifar-10","vgg")
    resnet_model, resnet_history,resnet_start_time, resnet_end_time = Train_model(resnet_model, data_gen, x_train_extra, y_train_extra, x_test_extra, y_test_extra,"cifar-10","resnet")
    googlenet_model, googlenet_history,googlenet_start_time, googlenet_end_time = Train_model(googlenet_model, data_gen, x_train_extra, y_train_extra, x_test_extra, y_test_extra,"cifar-10","googlenet")

    Save_classified_images(vgg_model, x_test_extra, y_test_extra, "cifar10", "vgg")
    Save_classified_images(resnet_model, x_test_extra, y_test_extra, "cifar10", "resnet")
    Save_classified_images(googlenet_model, x_test_extra, y_test_extra, "cifar10", "googlenet")


    # Save logs and visualize results
    Save_logs("cifar_vgg_logs.txt", vgg_history.history,vgg_start_time, vgg_end_time)
    Save_logs("cifar_resnet_logs.txt", resnet_history.history,resnet_start_time, resnet_end_time)
    Save_logs("cifar_googlenet_logs.txt", googlenet_history.history, googlenet_start_time, googlenet_end_time)

    Visualizing_Results(vgg_history, resnet_history, googlenet_history)
    return vgg_model, resnet_model, googlenet_model

#Define a function to evaluate the CIFAR-100 dataset
def Evaluate_cifar100():
    # Load and preprocess CIFAR-100 dataset
    (x_train, y_train), (x_test, y_test) = cifar100.load_data()

    # Normalize the pixel values
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

    data_gen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)
    y_train = to_categorical(y_train, 100)
    y_test = to_categorical(y_test, 100)

    # Define models
    vgg_model = Building_VGG_Model(input_shape=(32, 32, 3), num_classes=100)
    resnet_model = Building_Resnet_Model(input_shape=(32, 32, 3), num_classes=100)
    googlenet_model = Building_Googlenet_Model(input_shape=(32, 32, 3), num_classes=100)

    # Train and evaluate models
    vgg_model, vgg_history,vgg_start_time, vgg_end_time = Train_model(vgg_model, data_gen, x_train, y_train, x_test, y_test,"cifar-100","vgg")
    resnet_model, resnet_history,resnet_start_time, resnet_end_time = Train_model(resnet_model, data_gen, x_train, y_train, x_test, y_test,"cifar-100","resnet")
    googlenet_model, googlenet_history,googlenet_start_time, googlenet_end_time = Train_model(googlenet_model, data_gen, x_train, y_train, x_test, y_test,"cifar-100","googlenet")

    Save_classified_images(vgg_model, x_test, y_test, "cifar100", "vgg")
    Save_classified_images(resnet_model, x_test, y_test, "cifar100", "resnet")
    Save_classified_images(googlenet_model, x_test, y_test, "cifar100", "googlenet")

    # Save logs and visualize results
    Save_logs("cifar100_vgg_logs.txt", vgg_history.history,vgg_start_time, vgg_end_time)
    Save_logs("cifar100_resnet_logs.txt", resnet_history.history,resnet_start_time, resnet_end_time)
    Save_logs("cifar100_googlenet_logs.txt", googlenet_history.history, googlenet_start_time, googlenet_end_time)

    Visualizing_Results(vgg_history, resnet_history, googlenet_history)
    return vgg_model, resnet_model, googlenet_model




if __name__ == "__main__":
    vgg_model_mnist, resnet_model_mnist, googlenet_model_mnist = Evaluate_mnist()
    # Clear memory between dataset evaluations to prevent memory overflow
    keras.backend.clear_session()
    gc.collect()

    vgg_model_cifar10, resnet_model_cifar10, googlenet_model_cifar10 = Evaluate_cifar()
    # Repeat memory clearing
    keras.backend.clear_session()
    gc.collect()

    vgg_model_cifar100, resnet_model_cifar100, googlenet_model_cifar100 = Evaluate_cifar100()
    # Ensure memory is clear after all evaluations
    tf.keras.backend.clear_session()
    gc.collect()



# Load and preprocess MNIST data
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = mnist.load_data()
x_train_mnist, x_test_mnist = Data_preprocessing(x_train_mnist, x_test_mnist)
data_gen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)
y_train_mnist = to_categorical(y_train_mnist, 10)
y_test_mnist = to_categorical(y_test_mnist, 10)


# Load and preprocess CIFAR-10 data
x_train_extra, y_train_extra, x_test_extra, y_test_extra = Load_and_preprocess_extra_dataset()
data_gen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)

# Load and preprocess CIFAR-100 dataset
(x_train, y_train), (x_test, y_test) = cifar100.load_data()

# Normalize the pixel values
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

data_gen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)
y_train = to_categorical(y_train, 100)
y_test = to_categorical(y_test, 100)


def Compute_model_metrics(model_name):

    import numpy as np

    if model_name == "vgg":
        # Predict the labels for MNIST
        y_prediction_mnist = vgg_model_mnist.predict(x_test_mnist)
        y_prediction_mnist = np.argmax(y_prediction_mnist, axis=1)

        # Predict the labels for CIFAR-10
        y_prediction_cifar10 = vgg_model_cifar10.predict(x_test_extra)
        y_prediction_cifar10 = np.argmax(y_prediction_cifar10, axis=1)

        # Predict the labels for CIFAR-100
        y_prediction_cifar100 = vgg_model_cifar100.predict(x_test)
        y_prediction_cifar100 = np.argmax(y_prediction_cifar100, axis=1)

    if model_name == "resnet":

        # Predict the labels for MNIST
        y_prediction_mnist = resnet_model_mnist.predict(x_test_mnist)
        y_prediction_mnist = np.argmax(y_prediction_mnist, axis=1)

        # Predict the labels for CIFAR-10
        y_prediction_cifar10 = resnet_model_cifar10.predict(x_test_extra)
        y_prediction_cifar10 = np.argmax(y_prediction_cifar10, axis=1)

        # Predict the labels for CIFAR-100
        y_prediction_cifar100 = resnet_model_cifar100.predict(x_test)
        y_prediction_cifar100 = np.argmax(y_prediction_cifar100, axis=1)

    if model_name == "googlenet":

        # Predict the labels for MNIST
        y_prediction_mnist = googlenet_model_mnist.predict(x_test_mnist)
        y_prediction_mnist = np.argmax(y_prediction_mnist, axis=1)

        # Predict the labels for CIFAR-10
        y_prediction_cifar10 = googlenet_model_cifar10.predict(x_test_extra)
        y_prediction_cifar10 = np.argmax(y_prediction_cifar10, axis=1)

        # Predict the labels for CIFAR-100
        y_prediction_cifar100 = googlenet_model_cifar100.predict(x_test)
        y_prediction_cifar100 = np.argmax(y_prediction_cifar100, axis=1)



    y_true_mnist = y_test_mnist
    y_true_cifar10 = y_test_extra
    y_true_cifar100 = y_test

    def Extract_confusion_metrics_mnist(y_true, y_pred):
        if y_true.ndim > 1:
            y_true = np.argmax(y_true, axis=1)  # Convert one-hot encoded labels back to integers

        cm = confusion_matrix(y_true, y_pred)

        TP = np.diag(cm)
        FP = np.sum(cm, axis=0) - TP
        FN = np.sum(cm, axis=1) - TP
        TN = np.sum(cm) - (FP + FN + TP)

        return TP, FP, FN, TN


    def Extract_confusion_metrics_cifar(y_true, y_pred):
        if y_true.ndim > 1:
            y_true = np.argmax(y_true, axis=1)  # Convert one-hot encoded labels back to integers

        cm = confusion_matrix(y_true, y_pred)

        TP = np.diag(cm)
        FP = np.sum(cm, axis=0) - TP
        FN = np.sum(cm, axis=1) - TP
        TN = np.sum(cm) - (FP + FN + TP)

        return TP, FP, FN, TN

    TP0, FP0, FN0, TN0 = Extract_confusion_metrics_mnist(y_true_mnist, y_prediction_mnist)
    TP1, FP1, FN1, TN1 = Extract_confusion_metrics_cifar(y_true_cifar10, y_prediction_cifar10)
    TP2, FP2, FN2, TN2 = Extract_confusion_metrics_cifar(y_true_cifar100, y_prediction_cifar100)

    def precision(TP, FP):
        return TP / (TP + FP) if TP + FP > 0 else 0

    def recall(TP, FN):
        return TP / (TP + FN) if TP + FN > 0 else 0

    def f1_score(precision, recall):
        return 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0


    average_precision_mnist = np.mean([precision(tp, fp) for tp, fp in zip(TP0, FP0)])
    average_recall_mnist = np.mean([recall(tp, fn) for tp, fn in zip(TP0, FN0)])
    average_f1_score_mnist = np.mean([f1_score(precision(tp, fp), recall(tp, fn)) for tp, fp, fn in zip(TP0, FP0, FN0)])

    print("Average Precision for MNIST:", average_precision_mnist)
    print("Average Recall for MNIST:", average_recall_mnist)
    print("Average F1-score for MNIST:", average_f1_score_mnist)

    average_precision_cifar10 = np.mean([precision(tp, fp) for tp, fp in zip(TP1, FP1)])
    average_recall_cifar10 = np.mean([recall(tp, fn) for tp, fn in zip(TP1, FN1)])
    average_f1_score_cifar10 = np.mean([f1_score(precision(tp, fp), recall(tp, fn)) for tp, fp, fn in zip(TP1, FP1, FN1)])

    print("Average Precision for CIFAR-10:", average_precision_cifar10)
    print("Average Recall for CIFAR-10:", average_recall_cifar10)
    print("Average F1-score for CIFAR-10:", average_f1_score_cifar10)

    average_precision_cifar100 = np.mean([precision(tp, fp) for tp, fp in zip(TP2, FP2)])
    average_recall_cifar100 = np.mean([recall(tp, fn) for tp, fn in zip(TP2, FN2)])
    average_f1_score_cifar100 = np.mean([f1_score(precision(tp, fp), recall(tp, fn)) for tp, fp, fn in zip(TP2, FP2, FN2)])

    print("Average Precision for CIFAR-100:", average_precision_cifar100)
    print("Average Recall for CIFAR-100:", average_recall_cifar100)
    print("Average F1-score for CIFAR-100:", average_f1_score_cifar100)


Compute_model_metrics("vgg")

Compute_model_metrics("resnet")

Compute_model_metrics("googlenet")
```
